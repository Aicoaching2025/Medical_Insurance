```python
# Section 1: Scoping the Data & Exploratory Data Analysis

# Explore the dataset

def explore_data(df):
    """Basic exploration of the dataset."""
    print("\nFirst 5 rows:")
    print(df.head())
    print("\nDataset Info:")
    print(df.info())
    print("\nSummary Statistics:")
    print(df.describe())
    print("\nMissing Values:")
    print(df.isnull().sum())

# Visualization functions
def plot_distributions(df):
    """Plot distributions of numerical features."""
    df.hist(figsize=(12, 8), bins=30)
    plt.suptitle("Feature Distributions", fontsize=16)
    plt.show()

def plot_correlation_matrix(df):
    """Plot correlation heatmap."""
    # Select only numerical columns for correlation calculation
    numerical_df = df.select_dtypes(include=np.number)
    plt.figure(figsize=(10, 6))
    sns.heatmap(numerical_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("Feature Correlation Matrix")
    plt.show()
```

```python
explore_data(df)
```

```python
plot_distributions(df)
```

```python
plot_correlation_matrix(df)
```

```python
# Section 2: Build Functions for Each Analysis

def analyze_age_vs_charges(df):
    """Visualize the relationship between age and medical charges."""
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=df['age'], y=df['charges'], alpha=0.5)
    plt.xlabel("Age")
    plt.ylabel("Medical Charges")
    plt.title("Age vs. Medical Charges")
    plt.show()
```

```python
analyze_age_vs_charges(df)
```

```python
def analyze_bmi_vs_charges(df):
    """Visualize the relationship between BMI and medical charges."""
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=df['bmi'], y=df['charges'], alpha=0.5)
    plt.xlabel("BMI")
    plt.ylabel("Medical Charges")
    plt.title("BMI vs. Medical Charges")
    plt.show()
```

```python
analyze_bmi_vs_charges(df)
```

```python
def analyze_smoker_vs_charges(df):
    """Compare charges for smokers and non-smokers."""
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df['smoker'], y=df['charges'])
    plt.xlabel("Smoker")
    plt.ylabel("Medical Charges")
    plt.title("Smoker vs. Medical Charges")
    plt.show()
```

```python
analyze_smoker_vs_charges(df)
```
